# Visualization


The format of data which input into the function of visualization
```
[
    [[x,y of class1 data1],[x,y of class1 data2],[x,y of class1 data3]....[x,y of class1 data50]],
    [[x,y of class2 data1],[x,y of class2 data2],[x,y of class2 data3]....[x,y of class2 data50]],
    ...
    [[x,y of class20 data1],[x,y of class20 data2],[x,y of class20 data3]....[x,y of class20 data50]]
]
```

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand
import pickle
import numpy as np
from sklearn.manifold import TSNE
````

````python
labels = [
    'areoplane','bicycle','bird','boad','bottle',
    'bus','car','cat','chair','cow','dining table',
    'dog','horse','motobike','person','pottedplant',
    'sheep','sofa','train','tvmonitor']

colorbar = [
    '#e4007f', '#a40000','#ea68a2','#a84200','#f19149',
    '#fff45c','#8fc31f','#009944','#00736d','#0075a9',
    '#004986','#500047','#b28850','#81511c','#6a3906',
    '#59493f','#616e81','#898989','#89c997','#000000']


def visualisation(data):
    fig, ax = plt.subplots(figsize=(10, 10))
    i=0
    for color in colorbar:
        x = data[i][:,0]
        y = data[i][:,1]
        scale = 20
        ax.scatter(x, y, c=color, s=scale, label=labels[i],alpha=1, edgecolors='none')
        ax.legend()
        i=i+1
    plt.show()
````

## Second Result
### Loading the output of 
The `fc2_pre` is the output vector of fully connected of CNN while the `cap_vec` is the output of the Doc2Vec. The follwing code is to load the values of these two

````python
fc2_pre = pickle.load(open('fc2_pre','rb'))
cap_vec = pickle.load(open('cap_vec','rb'))
fc1_pre = pickle.load(open('fc1_pre_1000','rb'))
cap_vec.shape
````

### Combinition of the two vectors
Here we use a loop to UNION the two vectors

````python
# These two variables is the final combined feartus
captions_images_vectors = np.array([None]*1000)
captions_images_vectors_2 = np.array([None]*1000)

#UNION operation 1
for i in range(len(captions_images_vectors)):
    captions_images_vectors[i] = np.append(cap_vec[i],fc2_pre[i])
    
#UNION operation 2
for i in range(len(captions_images_vectors)):
    captions_images_vectors_2[i] = np.append(cap_vec[i],fc1_pre[i])
````

### Using t-SNE
In this part we build a instance of TSNE and use `fit_transform()` to reduce the dimension, and finally reshape it in order to visualize

````python
tsne = TSNE(perplexity=10, n_components=2, n_iter=5000,random_state=0)

# result 1
low_dim_embs = tsne.fit_transform(list(captions_images_vectors[:]))
low_dim_embs_rs = low_dim_embs.reshape((20,50,2))

#result 2
low_dim_embs_2 = tsne.fit_transform(list(captions_images_vectors_2[:]))
low_dim_embs_rs_2 = low_dim_embs_2.reshape((20,50,2))
````

````python
visualisation(low_dim_embs_rs)
````

````python
visualisation(low_dim_embs_rs_2)
````
